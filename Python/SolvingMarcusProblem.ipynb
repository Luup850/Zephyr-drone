```python
import pandas as pd

# Read in the log file
with open('../drone/mission/build/logs/Wed May  3 14_29_37 2023.log', 'rb') as f:
    log = f.read().decode('utf-8')
```

```python
# Remove the first 5 lines
log = log.split('\n')[5:]
log = '\n'.join(log)
```

```python
# Write to a csv
with open('./Wed-May-3-14_29_37-2023.csv', 'w') as f:
    f.write(log)
```

```python
# Open the csv as a dataframe
columns = ['Time', 'X', 'Y', 'Z', 'Pitch', 'Yaw', 'Roll', 'errorHeight', 'errorRoll', 'errorPitch', 'errorYaw']
print(len(columns))

# Create a data frame from the csv with named columns
df = pd.read_csv('./Wed-May-3-14_29_37-2023.csv', names=columns)


print(len(df))
df.head()
```

```python
#%matplotlib notebook

# Plot 3D the X,Y,Z columns
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
from matplotlib import cm
import numpy as np

fig = plt.figure(figsize = (12,12))
ax = fig.add_subplot(111, projection='3d')

ax.plot(df['X'], df['Y'], df['Z'])

ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('Z axis')

plt.show()
```
